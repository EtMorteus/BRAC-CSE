```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
from IPython.utils.io import temp_pyfile
class DoublyList:
  
  def __init__(self, a):
  # Creates a Non Dummy Headed Circular Doubly Linked List using the values from the given array a.
   
    if type(a)==list:
      self.head=Node(a[0],None,None)
      tail=self.head
      self.head.next=self.head
      self.head.prev=self.head
      
      for i in range(1,len(a)):
        temp=Node(a[i],None,None)
        tail.next=temp
        temp.prev=tail
        self.head.prev=temp
        temp.next=self.head
        tail=temp
    
    else:
     self.head=a
  
  # Counts the number of Nodes in the list and return the number
  def countNode(self):
   
    temp=self.head.next
    count=1
    while (temp!=self.head):
       count+=1
       temp=temp.next
    return count
  
  # prints the elements in the list
  def forwardprint(self):
    # To Do
    print(self.head.element, end=' ')
    
    temp=self.head.next
    while (temp!=self.head):
      print(temp.element, end=' ')
      temp=temp.next
    print()
      
      
      

  # prints the elements in the list backward
  def backwardprint(self):
    # To Do
    temp=self.head.prev
    while(temp!= self.head):
      print(temp.element, end=" ")
      temp=temp.prev
    print(self.head.element)
    

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    temp=self.head
    count=0
    while count!=idx:
      temp=temp.next
      count+=1


    if count==idx:
       return temp
    else:
      return None
      
  
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    temp=self.head
    count=0
    while temp.element!=elem:
      temp=temp.next
      count+=1
    if temp.element==elem:
       return count
    else:
      return -1
    

  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    n=Node(elem, None, None)
    if idx<0 or idx>self.countNode():
      return "Index invalid"
    
    else:
      temp=self.nodeAt(idx)
      n.next=temp
      n.prev=temp.prev
      temp.prev.next=n
      temp.prev=n
      return n
    
    
  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    n = Node(None,None,None)
    if idx<0 or idx>=self.countNode():
      return str(n.element)
      print('index error')
    else:
      temp= self.nodeAt(idx)
      temppp = temp.prev
      tempp = temp.next
      temppp.next = tempp
      tempp.prev = temppp    
      temp.next = temp.prev = None
      x = temp.element
      temp.element = None
      return str(x)
```
