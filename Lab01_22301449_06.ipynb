```python
# Complete the functions defined in this cell

# Test 01: Shift Left k cell
def shift_left(source, k):
  for i in range(len(source)):
    if i+k<len(source):
       source[i]=source[i+k]
    else:
       source[i]=0
  return source






# Test 02: Rotate Left k cell
def rotate_left(source, k):
    for j in range(k):
        temp=source[0]
        for i in range(len(source)-1):
            source[i]=source[i+1]
        source[-1]=temp
    return source




# Test 03: Shift Right k cell
def shift_right(a1,k):

    i=len(a1)-1 

    while(i>=k):
        a1[i]=a1[i-k] 
        i=i-1
    i=0
    
    while(i<k):
        a1[i]=0 
        i=i+1
    return a1





# Test 04: Rotate Right k cell
def rotate_right(source, k):
    n = 0
    while n < k:
        temp = source[0]
        i = 0
        while i <= len(source)-2:
            source[i] = source[i+1]
            i += 1
        source[len(source)-1] = temp
        n += 1
    return source

  # Hint, You can write a function for right rotate once and then use it



# Test 05: Remove an element from an array
def remove(source, idx):
  for i in source:
    i=source[idx]
  return source
  


# Test 06: Remove all occurrences of a particular element from an array
def remove_all(source, element):
    k = 0
    n = len(source)
    for i in range(n):
        if source[i] !=  element:
            source[k] = source[i]
            k += 1

    return source
  



# Test 07: Splitting an Array
def split_array(a):
    s = 0
    t = 0
    
    for i in a:
        t += i
    
    if t % 2 == 0:
        while s!= t /2:
            s += 1
    else:
        return False   
    if t / 2 == s:
        return True
    else:
        return False
    


# Test 08: Max Bunch Count
def max_bunch(a):
    
    m = 0
    count = 0
    i = 0
    
    for i in range(len(a)-1):
        if a[i+1] == a[i]:
            count += 1
            i += 1
        else:
            if count+1 > m:
                m = count+1
            count = 0
            i += 1
            
    if count+1 > m:
        m = count+1
        
    return m
```
